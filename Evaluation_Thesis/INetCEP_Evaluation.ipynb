## Interest Packets per second per query

```python
import pandas as pd
import seaborn as sns

sns.set_style("whitegrid")
interestData_data = pd.read_pickle('InterestPackets_Queries.pkl')
plotData = interestData_data[interestData_data['Approach']=="PRA"]
ax = sns.lineplot(data = plotData, x='Minute', y="Interest Packets", hue="Query Number")
#x = sns.pointplot(data = df, x='Sampling Rate', y="Recall", hue="Ratio interest / data packets", ci="sd" ,join=True)
ax.xaxis.set_tick_params(labelsize=15,rotation=45)
ax.set_xlabel("Time in Minutes", fontsize=20)
ax.yaxis.set_tick_params(labelsize=15)
ax.set_ylabel("Interest Packets / Minute", fontsize=20)
ax.legend(prop={'size':15})
#ax.set_yscale('log')
#ax.set_xscale('log')
figur = ax.get_figure()
figur.set_size_inches(8, 8)
#figur.savefig("InterestPackets_Queries_PRA.svg", bbox_inches='tight')
```

## Piechart Latency of Query

```python
import pandas as pd
import matplotlib.pyplot as plt
latency_data1 = pd.read_pickle('Pie_chart_data_Q3.pkl')
#latency_data2 = pd.read_pickle('Pie_chart_data_Q3_C.pkl')
import matplotlib as mpl
mpl.rcParams['font.size'] = 20
labels = []
for label in latency_data1['Type']:
    labels.append(label)
    
sizes1 = []
for size1 in latency_data1['Latency']:
    sizes1.append(size1)
    
#sizes2 = []
#for size2 in latency_data2['Latency']:
    #sizes2.append(size2)
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
#labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
#sizes = [15, 30, 45, 10]
#explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig, ax = plt.subplots()
colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink', 'tab:gray', 'tab:olive']

patches1, texts1, autotexts1 = ax.pie(sizes1, colors=colors, autopct='%1.1f%%', startangle=0)
#patches2, texts2, autotexts2 = axs[1].pie(sizes2, autopct='%1.1f%%', startangle=90)
#texts[0].set_fontsize(1)
ax.legend(patches1, labels, loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
ax.axis('equal')
ax.axis('equal')
#plt.tight_layout()
#plt.show()
#ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
figur = ax.get_figure()
#figur2 = axs[1].get_figure()
figur.set_size_inches(8,8)
#figur2.set_size_inches(8,8)
#fig.set_size_inches(24, 24)
#figur.savefig("piechart.svg", bbox_inches='tight')
```
