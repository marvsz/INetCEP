## Classes

```python
class Configuration:
    def __init__(self, configType, executionSpace, rate, run):
        self.configType = configType
        self.executionSpace = executionSpace
        self.rate = rate
        self.run = run
        self.actualRate = 0
        self.producers = 0

    def __repr__(self):
        return self.configType +" - "+ self.executionSpace +" - "+ self.rate
```

```python
class SecondStat:
    def __init__(self, timestamp, Interest_Packets, Data_Packets, PIT_Entries):
        self.timestamp = timestamp
        self.Interest_Packets = Interest_Packets
        self.Data_Packets = Data_Packets
        self.PIT_Entries = PIT_Entries
        
        def __repr__(self):
            return self.timestamp + "- Interest Packets: " + self.Interest_Packets + "Data Packets: " + self.Data_Packets + "PIT Entries: " + self.PIT_Entries
```

```python
class Tuple:
    def __init__(self, timestamp, data):
        self.timestamp = timestamp
        self.data = data
        
    def __repr__(self):
        return self.timestamp + ": " + self.data
```

```python
def get_last_n_lines(file_name, N):
    # Create an empty list to keep the track of last N lines
    list_of_lines = []
    # Open file for reading in binary mode
    with open(file_name, 'rb') as read_obj:
        # Move the cursor to the end of the file
        read_obj.seek(0, os.SEEK_END)
        # Create a buffer to keep the last read line
        buffer = bytearray()
        # Get the current position of pointer i.e eof
        pointer_location = read_obj.tell()
        # Loop till pointer reaches the top of the file
        while pointer_location >= 0:
            # Move the file pointer to the location pointed by pointer_location
            read_obj.seek(pointer_location)
            # Shift pointer location by -1
            pointer_location = pointer_location -1
            # read that byte / character
            new_byte = read_obj.read(1)
            # If the read byte is new line character then it means one line is read
            if new_byte == b'\n':
                # Save the line in list of lines
                list_of_lines.append(buffer.decode()[::-1])
                # If the size of list reaches N, then return the reversed list
                if len(list_of_lines) == N:
                    return list(reversed(list_of_lines))
                # Reinitialize the byte array to save next line
                buffer = bytearray()
            else:
                # If last read character is not eol then add it in buffer
                buffer.extend(new_byte)
 
        # As file is read completely, if there is still data in buffer, then its first line.
        if len(buffer) > 0:
            list_of_lines.append(buffer.decode()[::-1])
 
    # return the reversed list
    return list(reversed(list_of_lines))
```

```python
def parseNodeOutput(filename, data_sink):
    timestamp = 0
    intPack = 0
    datPack = 0
    pitEnt = 0
    stats = []
    with open(filename, 'r') as f:
        for line in f:
            if line.__contains__("ageing"):
                stats.append(SecondStat(timestamp,intPack,datPack,pitEnt))
                timestamp = int(line.split("=")[1].rstrip("\n\r"))
            if line.__contains__("Interest Packets recieved"):
                intPack = int(line.split(" ")[2].rstrip("\n\r"))
            if line.__contains__("Data Packets recieved"):
                datPack = int(line.split(" ")[2].rstrip("\n\r"))
            if line.__contains__("PIT Entries"):
                pitEnt = int(line.split(" ")[2].rstrip("\n\r"))
            if line.__contains__("Recieved Interest Packets"):
                data_sink['Recieved Interest Packets']=int(line.split(" ")[-1].rstrip("\n\r"))
            if line.__contains__("Recieved Persistent Interest Packets"):
                data_sink['Recieved Persistent Interest Packets']=int(line.split(" ")[-1].rstrip("\n\r"))
            if line.__contains__("Recieved Data Packets"):
                data_sink['Recieved Data Packets']=int(line.split(" ")[-1].rstrip("\n\r"))
            if line.__contains__("Recieved Data Stream Packets"):
                data_sink['Recieved Data Stream Packets']=int(line.split(" ")[-1].rstrip("\n\r"))
            if line.__contains__("Served Content Packets"):
                data_sink['Served Content Packets']=int(line.split(" ")[-1].rstrip("\n\r"))
            if line.__contains__("Overall Served Packets"):
                data_sink['Overall Served Packets']=int(line.split(" ")[-1].rstrip("\n\r"))
    data_sink['Stats']=stats
```

```python
def parseProducerOutput(filename, data_sink):
    with open(filename, 'r') as f:
        content = f.read()
        linesplit = content.split("[A]")
        for line in linesplit[1:]:
            lineContent = line.split(" ")
            data_sink['Tuple'].append(Tuple(lineContent[8],lineContent[5]))
        #print(linesplit)
```

```python
def parseConsumerOutput(filename, data_sink):
    with open(filename, 'r') as f:
        content = f.read()
        linesplit = content.split("[A]")
        for line in linesplit[1:]:
            time = line.split(" ")[9]
            packet = line.split(" Packet:")[1]
            data_sink['Tuple'].append(Tuple(time,packet))
```

```python
def countThroughput(filename):
    file = open(filename, 'r').read()
    stringToSearch = "Sendto"
    count = file.count(stringToSearch)
    return float(count)
```

```python
def calculateThroughput(sendMessages, Time):
    return sendMessages/Time
```

```python
def getConfiguration(configType, executionSpace, rate, configList):
    for config in configList:
        if((config.configType == configType) and (config.executionSpace == executionSpace) and (config.rate == rate)):
            return config
    return None
```

```python
def parseConfigurationList(configList):
    for approachDir in (PRALogDataLocation,UCLLogDataLocation):
        for directory in os.listdir(LogDataLocation+"/"+approachDir):
            dirString = LogDataLocation+"/"+approachDir+"/"+directory
            overallThroughput = 0
            fnString = ""
            producers = 0
            for subdir in os.listdir(dirString):
                subdirString = dirString + "/" + subdir
                throughput = 0
                for fn in os.listdir(subdirString):
                    fnString = subdirString + "/" + fn
                    #if(fn.startswith("producer")):
                    #    producers = producers + 1
                    #    throughput = throughput + calculateThroughput(countThroughput(fnString),float(get_last_n_lines(fnString,2)[0].split(" ")[1][:-1]))
                overallThroughput = overallThroughput + throughput
            configuration = parseConfig(fnString,"Userland")
            configuration.producers = producers
            configuration.actualRate = overallThroughput
            configList.append(configuration)
```

```python
import collections

def precision_score(actual, data):
    false_positives=len([item for item, count in collections.Counter(actual).items() if count > 1])
    false_negatives=len(list(set(data) - set(actual)))
    true_positives=len(data)-false_positives
    precision = true_positives/(true_positives+false_positives)
    return precision
```

```python
import collections

def recall_score(actual, data):
    false_positives=len([item for item, count in collections.Counter(actual).items() if count > 1])
    false_negatives=len(list(set(data) - set(actual)))
    true_positives=len(data)-false_positives
    recall = true_positives/(true_positives+false_negatives)
    return recall
```

```python
import collections

def f1_score(actual, data):
    false_positives=len([item for item, count in collections.Counter(actual).items() if count > 1])
    false_negatives=len(list(set(data) - set(actual)))
    true_positives=len(data)-false_positives
    precision = true_positives/(true_positives+false_positives)
    recall = true_positives/(true_positives+false_negatives)
    f1 = 2*(precision*recall)/(precision+recall)
    return f1
```

```python
import collections

def accuracy_score(actual, data):
    false_positives=len([item for item, count in collections.Counter(actual).items() if count > 1])
    false_negatives=len(list(set(data) - set(actual)))
    true_positives=len(data)-false_positives
    accuracy = true_positives/(true_positives+false_positives+false_negatives)
    return accuracy
```

```python
def generatePrecisionData(filename):
    data_frame = pd.read_pickle(filename)
    AccuracyData = {'Run': [],'Config Type':[], 'Execution Space':[], 'Rate':[],
                'Accuracy':[], 'Precision': [], 'Recall': [], 'F1 Score': []}
    for i in range(0,len(data_frame["Run"])):
        AccuracyData['Run'].append(data_frame['Run'][i])
        AccuracyData['Config Type'].append(data_frame['Config Type'][i])
        AccuracyData['Execution Space'].append(data_frame['Execution Space'][i])
        AccuracyData['Rate'].append(data_frame['Rate'][i])
        expected = data_frame["Expected"][i]
        actual = data_frame["Actual"][i]
        expected.sort()
        actual.sort()
        AccuracyData['Accuracy'].append(accuracy_score(actual,expected))
        AccuracyData['Precision'].append(precision_score(actual,expected))
        AccuracyData['Recall'].append(recall_score(actual,expected))
        AccuracyData['F1 Score'].append(f1_score(actual,expected))
    df = pd.DataFrame(AccuracyData)
    df.to_pickle('accuracy_'+filename)
```

## Precision, Recall, F-Score and Accuracy Data Generator
We split up the different settings into their own pkls since we did not have enough RAM.
For each data Frame a the Gold Standard are the packets the producer produced.

```python
import pandas as pd
prod_data_frame = pd.read_pickle('producer_data.pkl')
con_data_frame = pd.read_pickle('consumer_data.pkl')
const_con_data_frame = pd.read_pickle('constant_consumer_data.pkl')
frames = [con_data_frame,const_con_data_frame]
combined_consumer_data_frame = pd.concat(frames)


executionSpaces = prod_data_frame['Execution Space'].unique()
rates = prod_data_frame['Rate'].unique()
configs = prod_data_frame['Config Type'].unique()
for e in executionSpaces:
    is_e =  prod_data_frame['Execution Space'] == e
    executionSpaceFilteredDataFrame = prod_data_frame[is_e]
    cons_is_e = combined_consumer_data_frame['Execution Space'] == e
    cons_executionSpaceFilteredDataFrame = combined_consumer_data_frame[cons_is_e]
    
    for r in rates:
        is_r = executionSpaceFilteredDataFrame['Rate'] == r
        rateFilteredDataFrame = executionSpaceFilteredDataFrame[is_r]
        cons_is_r = cons_executionSpaceFilteredDataFrame['Rate'] == r
        cons_rateFilteredDataFrame = cons_executionSpaceFilteredDataFrame[cons_is_r]
        
        for c in configs:
            is_c = rateFilteredDataFrame['Config Type'] == c
            configFilteredDataFrame = rateFilteredDataFrame[is_c]
            cons_is_c = cons_rateFilteredDataFrame['Config Type'] == c
            cons_configFilteredDataFrame = cons_rateFilteredDataFrame[cons_is_c]
            
            runs = configFilteredDataFrame['Run'].unique()
            Data={'Run':[],'Config Type':[],'Execution Space':[],'Rate':[],'Expected':[], 'Actual':[]}
            for run in runs:
                is_run = configFilteredDataFrame['Run'] == run
                runFilteredDataFrame = configFilteredDataFrame[is_run]
                
                cons_is_run = cons_configFilteredDataFrame['Run'] == run
                cons_runFilteredDataFrame = cons_configFilteredDataFrame[cons_is_run]
                expected = []
                for i in range(0,len(runFilteredDataFrame)):
                    for t in runFilteredDataFrame.iloc[i]['Tuple']:
                        splitDat = t.data.split("/")
                        resVal = splitDat[1]+"/"+splitDat[2] 
                        expected.append(resVal)
                actual = []
                for i in range(0,len(cons_runFilteredDataFrame)):
                    for t in cons_runFilteredDataFrame.iloc[i]['Tuple']:
                        splitDat = t.data.split("/")
                        resVal = splitDat[1]+"/"+splitDat[2] 
                        actual.append(resVal)
                Data['Run'].append(run)
                Data['Config Type'].append(c)
                Data['Execution Space'].append(e)
                Data['Rate'].append(r)
                Data['Expected'].append(expected)
                Data['Actual'].append(actual)
            fileName = "data"+c+e+r
            df = pd.DataFrame(Data)
            df.to_pickle(fileName+'.pkl')
```
