# Plotting

## Precision Windows

```python
import seaborn as sns
import pandas as pd

df = pd.read_excel (r'EvalTable.xlsx', sheet_name='Window Implementation Comparison')
df['Hue'] = df['Implementation'] + ", " + df['Execution Space'] + ", " + df['State location']
sns.set_style("whitegrid")
df['Precision'] = df['Precision'] / 100
ax = sns.pointplot(data = df, x='# of Tuples', y="Precision", hue="Hue", ci="sd" ,join=False)
ax.xaxis.set_tick_params(labelsize=15,rotation=45)
ax.set_xlabel("Number of Tuples in the Window", fontsize=20)
ax.yaxis.set_tick_params(labelsize=15)
ax.set_ylabel("Precision", fontsize=20)
ax.legend(prop={'size':15})
#ax.set_yscale('log')
#ax.set_xscale('log')
figur = ax.get_figure()
figur.set_size_inches(8, 4)
#figur.savefig("Precision_Windows.svg", bbox_inches='tight')
```

## Latency Windows

```python
import seaborn as sns
import pandas as pd

df = pd.read_excel (r'EvalTable.xlsx', sheet_name='Window Implementation Comparison')
df['Hue'] = df['Implementation'] + ", " + df['Execution Space'] + ", " + df['State location']
df['Latency'] = df['Latency'] *1000
sns.set_style("whitegrid")
ax = sns.pointplot(data = df, x='# of Tuples', y="Latency", hue="Hue", ci="sd" ,join=False)
ax.xaxis.set_tick_params(labelsize=15,rotation=45)
ax.set_xlabel("Number of Tuples in the Window", fontsize=20)
ax.yaxis.set_tick_params(labelsize=15)
ax.set_ylabel("Latency in Milliseconds", fontsize=20)
ax.legend(prop={'size':15})
#ax.set_yscale('log')
#ax.set_xscale('log')
figur = ax.get_figure()
figur.set_size_inches(8, 4)
#figur.savefig("Latency_Windows.svg", bbox_inches='tight')
```

## Concurrent Windows

```python
import pandas as pd
import seaborn as sns

sns.set_style("whitegrid")
latency_data = pd.read_pickle('Latency_ConcurrentWindows.pkl')
latency_data['Hue'] = latency_data['Config Type'] + ", " + latency_data['Execution Space']
latency_data['Latency'] = latency_data['Latency'] * 1000
ax = sns.pointplot(data = latency_data, x='Concurrent Windows', y="Latency", hue="Hue",join=True)
ax.xaxis.set_tick_params(labelsize=15,rotation=45)
ax.set_xlabel("Number of Concurrent Windows", fontsize=20)
ax.yaxis.set_tick_params(labelsize=15)
ax.set_ylabel("Latency in Milliseconds", fontsize=20)
ax.legend(prop={'size':15})
#ax.set_yscale('log')
#ax.set_xscale('log')
figur = ax.get_figure()
figur.set_size_inches(8, 8)
#figur.savefig("Latency_ConcurrentWindows.svg", bbox_inches='tight')
```
